```python
import pandas as pd

# PASO 1: CREAR EL TRADUCTOR ISO -> NOMBRE 
diccionario_iso = {
    'ARG': 'Argentina', 'BHS': 'Bahamas', 'BRB': 'Barbados', 'BLZ': 'Belice',
    'BOL': 'Bolivia', 'BRA': 'Brasil', 'CHL': 'Chile', 'COL': 'Colombia',
    'CRI': 'Costa Rica', 'CUB': 'Cuba', 'DOM': 'República Dominicana',
    'ECU': 'Ecuador', 'SLV': 'El Salvador', 'GRD': 'Granada', 'GTM': 'Guatemala',
    'GUY': 'Guyana', 'HTI': 'Haití', 'HND': 'Honduras', 'JAM': 'Jamaica',
    'MEX': 'México', 'NIC': 'Nicaragua', 'PAN': 'Panamá', 'PRY': 'Paraguay',
    'PER': 'Perú', 'LCA': 'Santa Lucía', 'VCT': 'San Vicente y las Granadinas',
    'SUR': 'Suriname', 'TTO': 'Trinidad y Tabago', 'URY': 'Uruguay', 'VEN': 'Venezuela'
}

# PASO 2: NORMALIZAR NOMBRES DE COLUMNAS
# Tabla 1: Matriculación
matriculacionF_norm = matriculacionF.copy()
# Traducimos el código ISO al nombre completo
matriculacionF_norm['Pais'] = matriculacionF_norm['geoUnit'].map(diccionario_iso)
matriculacionF_norm = matriculacionF_norm.rename(columns={'year': 'Anio', 'famale': 'Matricula_Mujeres', 'male': 'Matricula_Hombres'})
# Nos quedamos solo con las columnas útiles 
matriculacionF_norm = matriculacionF_norm[['Pais', 'Anio', 'Matricula_Mujeres', 'Matricula_Hombres']].dropna(subset=['Pais'])

# Tabla 2: PBI
pbiF_norm = pbiF.copy()
pbiF_norm = pbiF_norm.rename(columns={'Anios': 'Anio'})

# Tabla 3: Puestos Directivos
puestosDirF_norm = puestosDirF.copy()
puestosDirF_norm = puestosDirF_norm[['Pais', 'Anio', 'Porcentaje_Directivas']]

# Tabla 4: Población
poblacionF_norm = poblacionF.copy()
poblacionF_norm = poblacionF_norm.rename(columns={'Anios': 'Anio'})

pob_total = poblacionF_norm.groupby(['Pais', 'Anio'])['CantidadPoblacion'].sum().reset_index()


# PASO 3: (MERGE) 
# Unimos Matriculación con Población
df_final = pd.merge(matriculacionF_norm, pob_total, on=['Pais', 'Anio'], how='outer')

# Unimos con PBI
df_final = pd.merge(df_final, pbiF_norm, on=['Pais', 'Anio'], how='outer')

# Unimos con Puestos Directivos
df_final = pd.merge(df_final, puestosDirF_norm, on=['Pais', 'Anio'], how='outer')

# PASO 4: LIMPIEZA 
# 
df_final = df_final.sort_values(by=['Pais', 'Anio']).reset_index(drop=True)

# Veamos el resultado
df_final.info()
```

```python
from sklearn.linear_model import Ridge
import pandas as pd
import numpy as np

print("Iniciando la simulación del futuro (hasta 2100)...")

# 1. Definimos las variables de entrada 
features = ['Anio', 'CantidadPoblacion']
targets = ['Matricula_Mujeres', 'tasaCrecimiento', 'Porcentaje_Directivas']

# Lista para guardar los datos procesados de todos los países
predicciones_totales = []

# 2. Obtenemos la lista de países únicos
paises = df_final['Pais'].dropna().unique()

for pais in paises:
    # Aislamos los datos del país actual
    df_pais = df_final[df_final['Pais'] == pais].copy()
    
    # Separamos el PASADO (datos históricos reales para entrenar)
    df_train = df_pais.dropna(subset=['Matricula_Mujeres', 'Porcentaje_Directivas', 'tasaCrecimiento'])
    
    # Separamos el FUTURO (años donde tenemos población proyectada pero nos faltan los porcentajes)
    df_future = df_pais[df_pais['Matricula_Mujeres'].isna() & df_pais['CantidadPoblacion'].notna()].copy()
    
    # Si un país no tiene suficientes datos históricos (mínimo 3 años), lo guardamos como está y lo saltamos
    if len(df_train) < 3 or len(df_future) == 0:
        # Etiquetamos lo que haya como Histórico
        if len(df_train) > 0:
            df_pais['Tipo_Dato'] = 'Histórico'
        predicciones_totales.append(df_pais) 
        continue
        
    # --- ENTRENAMIENTO Y PREDICCIÓN 
    for target in targets:
        # Iniciamos el motor Lineal con Regularización (Ridge)
        ridge_model = Ridge(alpha=1.0) 
        
    
        X_train = df_train[features]
        y_train = df_train[target]
        
        # Entrenamos el modelo 
        ridge_model.fit(X_train, y_train)
        
        # Predecimos el futuro
        X_future = df_future[features]
        prediccion_cruda = ridge_model.predict(X_future)
        
        # Limitamos porcentajes entre 0 y 100
        if target in ['Matricula_Mujeres', 'Porcentaje_Directivas']:
            df_future[target] = np.clip(prediccion_cruda, 0, 100)
        else:
            # Para la tasa de crecimiento 
            df_future[target] = prediccion_cruda
    
    # Calculamos a los Hombres automáticamente 
    df_future['Matricula_Hombres'] = 100 - df_future['Matricula_Mujeres']
    
    # Etiquetamos visualmente qué es real y qué es simulado
    df_train['Tipo_Dato'] = 'Histórico'
    df_future['Tipo_Dato'] = 'Predicción'
    
    # Unimos el pasado real con el futuro predicho para este país
    df_pais_completo = pd.concat([df_train, df_future]).sort_values('Anio')
    
    # Lo agregamos a la lista maestra
    predicciones_totales.append(df_pais_completo)


df_prediccion_final = pd.concat(predicciones_totales)

# Redondeamos los números
df_prediccion_final = df_prediccion_final.round(2)

# Exportamos el JSON
nombre_archivo = 'prediccion_filtro_invisible_2100.json'
df_prediccion_final.to_json(nombre_archivo, orient='records')

print(f"¡Simulación terminada con éxito! Archivo '{nombre_archivo}' generado y listo para React.")
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import Ridge
import pandas as pd

# 1. Creamos una lista 
resultados_evaluacion = []

# 2. Iteramos sobre todos los países
paises = df_final['Pais'].dropna().unique()
features = ['Anio', 'CantidadPoblacion']

for pais in paises:
    # Aislamos el pasado histórico del país
    df_pais = df_final[df_final['Pais'] == pais].dropna(subset=['Matricula_Mujeres', 'Porcentaje_Directivas'])
    
    # Exigimos al menos 3 años de datos para que la métrica 
    if len(df_pais) < 3:
        continue
        
    X_real = df_pais[features]
    y_matricula = df_pais['Matricula_Mujeres']
    y_directivas = df_pais['Porcentaje_Directivas']
    
    # 3. Entrenamos modelos rápidos para evaluar
    modelo_mat = Ridge(alpha=1.0).fit(X_real, y_matricula)
    modelo_dir = Ridge(alpha=1.0).fit(X_real, y_directivas)
    
    pred_mat = modelo_mat.predict(X_real)
    pred_dir = modelo_dir.predict(X_real)
    
    # 4. Calculamos las métricas 
    r2_mat = r2_score(y_matricula, pred_mat)
    mae_mat = mean_absolute_error(y_matricula, pred_mat)
    
    r2_dir = r2_score(y_directivas, pred_dir)
    mae_dir = mean_absolute_error(y_directivas, pred_dir)
    
    # Guardamos los resultados
    resultados_evaluacion.append({
        'Pais': pais,
        'R2_Matricula': round(r2_mat, 2),
        'MAE_Mat_Puntos_Porcentuales': round(mae_mat, 2),
        'R2_Directivas': round(r2_dir, 2),
        'MAE_Dir_Puntos_Porcentuales': round(mae_dir, 2)
    })

# 5. Convertimos a tabla y la ordenamos 
df_metricas = pd.DataFrame(resultados_evaluacion)
df_metricas = df_metricas.sort_values(by='R2_Matricula', ascending=False).reset_index(drop=True)

# Imprimimos la tabla
print("--- Resultado del Modelo Ridge por País ---")
print(df_metricas.to_string())
```
